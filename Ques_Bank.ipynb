```python
import pandas as pd
from random import shuffle,random
from random import choice
import json
# reading csv file from url 
dict_Shuffle_Ques = []
dict_Shuffle_Ans = []
Q_dict={}
Q_bank=[]
data = pd.read_excel('C:\\Users\\Ultimate\\Desktop\\Q_BANK.xlsx').to_dict(orient='index')

for k,v in data.items():
    dict_Shuffle_Ques.append(k)
    dict_Shuffle_Ans.append(v)
shuffle(dict_Shuffle_Ques)
shuffle(dict_Shuffle_Ans)
zipbObj = zip(dict_Shuffle_Ques, dict_Shuffle_Ans)
dictOfWords = dict(zipbObj)

for k,v in dictOfWords.items():
    if k in range(0,5):
        list1=[dictOfWords[k]['A'],dictOfWords[k]['B'],dictOfWords[k]['C'],dictOfWords[k]['D']]
        Q_dict['Question'] = [dictOfWords[k]['Question']]
        Q_dict['Answer']=list1
        Q_dict['Answer_Key']=[dictOfWords[k]['Answer Choice']]
        Q_dict['Answer_Key']=[dictOfWords[k]['Answer']]
        #print(Q_dict)
        for k,v in Q_dict.items():
            l= [i for i in v] #Creates a list
            shuffle(l) #shuffles the list.
            Q_dict[k]=l
        Q_bank.append(Q_dict)
print(json.dumps(Q_bank,indent=4))
```
