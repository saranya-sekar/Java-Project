```python
from collections import defaultdict
weather_dict = {} 
dict_key_count = 0 
while True: 
    User = str(input("Please Enter Your Role :: ")) 
    print(User) 
    if User == '': 
        print("Please Enter Valid Role .. ") 
        continue 
        
## Data Entry
    if User.casefold() == 'Data Entry'.casefold():
        while True:
            City = input("Please Enter City :: ")
            if City =="":
                if weather_dict == '':
                    continue
                elif weather_dict:
                    break

            while True:
                Location = input("Please Enter Location :: ")
                if Location =="":
                    if weather_dict == '':
                        continue
                    elif weather_dict:
                        break
                break
            while True:
                try:
                # Precipitation,Humidity,Wind,Temperature = input("Enter weather data Precipitation, Humidity , Wind and Temperature separated by space:").split()  
                    Precipitation =input("Enter weather data Precipitation ::")
                    Humidity = input("Enter weather data Humidity ::")
                    Wind = input("Enter weather data Wind ::")
                    Temperature = input("Enter weather data Temperature ::") 
                except ValueError:
                    print("That wasn't a valid entry!")
                    break
                if weather_dict == '':
                    continue
                elif weather_dict:
                    break
                break    
            weather_dict[City] = []
            if Location == '':
                break
            elif Location !='' and Precipitation != '' and Humidity !='' and Wind!='' and Temperature !='':   
                weather_dict[City].append(Location)    
                weather_dict[City].append(Precipitation+"%")
                weather_dict[City].append(Humidity+"%")
                weather_dict[City].append(Wind+"km/h")
                weather_dict[City].append(Temperature+" C")
                list_value = weather_dict[City]
                list_value[4] = str(Temperature+" C")+" | "+ str((int(Temperature) * 1.8) + 32) + 'F'
                weather_dict[City] = list_value    
                print("Complete Weather Report For View :: \n",weather_dict)   
                all_pairs = list(weather_dict.items())
                if weather_dict == '':
                    break
                if weather_dict !=''and weather_dict[City] !='':
                    try:
                        if all_pairs[dict_key_count] == '':
                            break
                        elif all_pairs[dict_key_count]:    
                            print("Weather Record for City {}".format(City)+ " is :: {}".format(all_pairs[dict_key_count]))
                            dict_key_count +=1
                    except IndexError:
                        print("City already in the data.. Please enter some other city..")
                        break
                    view_all = str(input("Do you want to view any other city record.. yes or no "))
                    if view_all =='' or view_all =='no'.casefold():
                        break 
                    if view_all == 'yes'.casefold():
                        while True:
                            view_city = str(input("Enter City Name To Get Weather Report:: "))
                            if view_city =='' or view_city =='no'.casefold():
                                break
                            if view_city !='':
                                Result = {k:v for (k,v) in weather_dict.items() if (City.casefold() == k.casefold() )}
                                print("Complete Weather Report For View for City {}:: ".format(City),Result)
                    break

            continue

  ##admin
    elif User.casefold() == 'Admin'.casefold():
        while True:
            City = str(input("Please Enter City :: ")).casefold()
            weather_update = str(input("Please Enter Field (Precipitation,Humidity,Wind,Temperature)  :: "))
            New_Value = int(input("Please Enter New Value For :: {}".format(weather_update)))
            if City=="" or weather_update=="" or New_Value == "":
                print("That wasn't a valid entry!") 
                continue   
            if weather_update.casefold() == 'Precipitation'.casefold():
                list_value = weather_dict[City]
                list_value[1] = New_Value + '%'
                weather_dict[City] = list_value
            if weather_update.casefold() == 'Humidity'.casefold():
                list_value = weather_dict[City]
                list_value[2] = New_Value + '%'
                weather_dict[City] = list_value
                print(weather_dict) 
            if weather_update.casefold() == 'Wind'.casefold():
                list_value = weather_dict[City]
                list_value[3] = New_Value + 'km/h'
                weather_dict[City] = list_value
            if weather_update.casefold() == 'Temperature'.casefold():
                list_value = weather_dict[City]
                list_value[4] = str(New_Value+" C")+" | "+ str((int(New_Value) * 1.8) + 32) + 'F'
                weather_dict[City] = list_value 
            print("Complete Updated Weather Report For View :: \n",weather_dict)
    ##End User
    if User.casefold() == 'User'.casefold():
        while True:
            City = str(input("Please Enter City For Getting Weather Report:: "))
            if City =='':
                print("Please enter any City")
                break  
            if City in weather_dict.keys():    
                Result = {k:v for (k,v) in weather_dict.items() if (City.casefold() == k.casefold() )}
                print("Complete Weather Report For View For City {}:: \n".format(City),Result)
            elif City not in weather_dict.keys():
                print("City is Not in the Data..")
             
        continue        
```
