```python
import csv
import json

filename_Disney_Hotstar="C:\\Users\\Ultimate\\Desktop\\IMDB\\Disney_Hotstar.csv"
filename_Netflix ="C:\\Users\\Ultimate\\Desktop\\IMDB\\Netflix.csv"
filename_Amazon_Prime_Video="C:\\Users\\Ultimate\\Desktop\\IMDB\\Amazon_Prime_Video.csv"
filename_zee5 ="C:\\Users\\Ultimate\\Desktop\\IMDB\\zee5.csv"
filename_Sony_Liv="C:\\Users\\Ultimate\\Desktop\\IMDB\\Sony_Liv.csv"
filename_Voot="C:\\Users\\Ultimate\\Desktop\\IMDB\\Voot.csv"
filename_AltBalaji="C:\\Users\\Ultimate\\Desktop\\IMDB\\AltBalaji.csv"
filename_JioCinema="C:\\Users\\Ultimate\\Desktop\\IMDB\\JioCinema.csv"
filename_MX_Player="C:\\Users\\Ultimate\\Desktop\\IMDB\\MX_Player.csv"
filename_YouTube_Premium="C:\\Users\\Ultimate\\Desktop\\IMDB\\YouTube_Premium.csv"

Disney_Hotstar = []
Netflix = []
Amazon_Prime_Video = []
zee5 = []
Sony_Liv = []
Voot = []
AltBalaji = []
JioCinema = []
MX_Player = []
YouTube_Premium = []


ott_dict_copy={}
ott_dict = {}
ott_list =[]
ott_full_list = []

your_role = str(input("Enter your role: "))

def csv_toDict(filename,ott_name):
    with open(filename, 'r') as data:
        
        lines = csv.DictReader(data)
        for line in lines:
            ott_name.append(line)
        dict_ = {indic['movie_name']:{ k:indic[k] for k in indic if k != 'name'} for indic in ott_name}
        return dict_ 
 

Disney_Hotstar_movies_dict = csv_toDict(filename_Disney_Hotstar,Disney_Hotstar)
Netflix_movies_dict = csv_toDict(filename_Netflix,Netflix) 
Amazon_Prime_Video_movies_dict = csv_toDict(filename_Amazon_Prime_Video,Amazon_Prime_Video)
zee5_movies_dict = csv_toDict(filename_zee5,zee5)
Sony_Liv_movies_dict = csv_toDict(filename_Sony_Liv,Sony_Liv)
Voot_movies_dict = csv_toDict(filename_Voot,Voot)
AltBalaji_movies_dict  = csv_toDict(filename_AltBalaji,AltBalaji)
JioCinema_movies_dict = csv_toDict(filename_JioCinema,JioCinema)
MX_Player_movies_dict = csv_toDict(filename_MX_Player,MX_Player)
YouTube_Premium_movies_dict = csv_toDict(filename_YouTube_Premium,YouTube_Premium)

ott_name_list = ['Disney+ Hotstar','Netflix','Amazon Prime Video','zee5','Sony Liv','Voot','AltBalaji','JioCinema','MX Player','YouTube Premium']
ott_name_dict_list = {'Disney+ Hotstar':Disney_Hotstar_movies_dict,'Netflix':Netflix_movies_dict,'Amazon Prime Video':Amazon_Prime_Video_movies_dict,'zee5':zee5_movies_dict,'Sony Liv':Sony_Liv_movies_dict,'Voot':Voot_movies_dict,'AltBalaji':AltBalaji_movies_dict,'JioCinema':JioCinema_movies_dict,'MX Player':MX_Player_movies_dict,'YouTube Premium':YouTube_Premium_movies_dict}



def create_dict(ott_key,ott_list):
    ott_dict[ott_key] = ott_list
    return ott_dict

def ott_full_list_util(ott_name_dict_list,ott_full_list):
    for k,v in ott_name_dict_list.items():
        ott_full_list.append(create_dict(k,ott_name_dict_list[k]))
    return ott_full_list

   
def search_movie(movie_name_search,ott_name,ott_full_dict,ott_name_dict_list):
    search_result = ''
    global_search_result =[]
    dictOfSearch ={}
    ott_found =[]
    for k,v in ott_name_dict_list.items():
        if(ott_name == k):
            for item in ott_full_dict :
                for ott_name1 in ott_name_list:
                    try:
                        if item[ott_name1][movie_name_search]!= "":
                            search_result = item[ott_name1][movie_name_search]
                            if search_result != '' and ott_name1 ==ott_name:
                                global_search_result.append(item[ott_name1][movie_name_search])
                                ott_found.append(ott_name1)
                    except KeyError:
                        search_result =''
                    if search_result != '' and ott_name1 ==ott_name:       
                        zipbObj = zip(ott_found, global_search_result)
                        dictOfSearch = dict(zipbObj) 
    if dictOfSearch !='':
        return ott_name,dictOfSearch


    

def search_full_ott(movie_name_search,ott_full_dict,ott_name_list): 
    global_search_result =[]
    dictOfSearch ={}
    ott_found =[]
    for item in ott_full_dict :
        for ott_name in ott_name_list:
            try:
                if item[ott_name][movie_name_search] !="":
                    search_result = item[ott_name][movie_name_search]
                    global_search_result.append(item[ott_name][movie_name_search])
                    ott_found.append(ott_name)
            except KeyError:
                search_result =''
            if search_result != '':       
                zipbObj = zip(ott_found, global_search_result)
                dictOfSearch = dict(zipbObj)
    if dictOfSearch !='':
        return dictOfSearch

def create_update_data(ott_name_dict_list,ott_to_update,movie_name_add,movie_year_add,movie_desc_add,imdb_rating_add,Stars_add,Director_add,Writers_add):
    for key,val in ott_name_dict_list.items():
        if key == ott_to_update:
            val[movie_name_add] = {'movie_name':movie_name_add,'movie_year':movie_year_add,'movie_desc':movie_desc_add,'imdb_rating':imdb_rating_add,'Stars':Stars_add,'Director':Director_add,'Writers':Writers_add}
            return ott_to_update,val


ott_full_dict = ott_full_list_util(ott_name_dict_list,ott_full_list)

if your_role == 'user' or your_role =='admin':
    movie_find_global = {} 
    movie_find={}
    ott_name_global = ''
    ott_name_found=''
    movie_name_search = str(input("\n Enter Movie name to Search :: "))
    ott_name = str(input("\n Enter OTT Name :: \n Disney+ Hotstar \n Netflix  \n Amazon Prime Video \n zee5 \n Sony Liv \n Voot \n AltBalaji \n JioCinema  \n MX Player \n YouTube Premium  ::\n"))
    #count =0
    try:
        ott_name_found,movie_find = search_movie(movie_name_search,ott_name,ott_full_dict,ott_name_dict_list)
        if movie_find !={}:
            print("Found Movie in ",ott_name_found,"\n movie data ::\n",json.dumps(movie_find,indent=4))
        if movie_find =={}:
            print("Not Found Movie in ",ott_name_found,"\n movie data ::\n",json.dumps(movie_find,indent=4))
            movie_find_global = search_full_ott(movie_name_search,ott_full_list,ott_name_list)
            if movie_find_global!= {}:
                for k,v in movie_find_global.items():
                    count+=1
                if count == 1:
                    print("Not Found in ",ott_name," But Found Movie in Global Search :: \n",json.dumps(movie_find_global,indent=4))
    except ValueError:
        print("\n\n =========== No movies found in :: ",ott_name)
    except TypeError:
        print("\n\n =========== No movies found in :: ",ott_name)
    if movie_find == {}:
        movie_find_global = search_full_ott(movie_name_search,ott_full_list,ott_name_list)
        if movie_find_global!= {}:
            for k,v in movie_find_global.items():
                count+=1
            if count > 1:
                 print("Not Found in ",ott_name," \n But Found Movie in Global Search :: \n",json.dumps(movie_find_global,indent=4))
          
            
if your_role =='data entry' or your_role =='admin':
    ott_to_update = str(input('Enter the OTT name to Update::\n \n Disney+ Hotstar \n Netflix  \n Amazon Prime Video \n zee5 \n Sony Liv \n Voot \n AltBalaji \n JioCinema  \n MX Player \n YouTube Premium \n'))
    movie_name_add = str(input('Enter movie name to add:: \n'))
    movie_year_add = str(input('Enter movie year to add:: \n'))
    movie_desc_add = str(input('Enter movie desc to add:: \n'))
    imdb_rating_add = str(input('Enter movie imdb rating to add:: \n'))
    Stars_add = str(input('Enter Stars name to add:: \n'))
    Director_add = str(input('Enter movie Director to add:: \n'))
    Writers_add = str(input('Enter movie Writers to add:: \n'))
    
    ott_to_update1, data_update = create_update_data(ott_name_dict_list,ott_to_update,movie_name_add,movie_year_add,movie_desc_add,imdb_rating_add,Stars_add,Director_add,Writers_add)
    print("ott_to_update1::\n",ott_to_update1)
    print("data_update::\n",data_update)

```
